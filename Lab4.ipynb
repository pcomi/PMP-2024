Consideram o retea Markov de dimensiune mica care descrie distributia probabilitatilor pentru 5 variabile discrete binare: A1, A2, A3, A4 si A5.
Relatiile de independenta conditionata dintre aceste variabile sunt descrise de urmatoarele conexiuni:
A1 conectat cu A2 si A3, A2 este conectat cu A4 si A5, A3 este conectat cu A4, iar A4 este conectat cu A5.
Se cunoaste urmatoarea distributie a energiei pentru clicile modelului:
fi de Ai1, Ai2, ... Aik = e ^ (i1Ai1 + i2Ai2 + ... + ikAik)

a) Vizualizati in pgmpy graful retelei Markov, iar apoi determinati clicile modelului

```python
import networkx as nx
import matplotlib.pyplot as plot

model = nx.Graph()

edges = [("A1", "A2"), ("A1", "A3"), ("A2", "A4"), ("A2", "A5"), ("A3", "A4"), ("A4", "A5")]
model.add_edges_from(edges)

pos = nx.spring_layout(model)

nx.draw(model, pos, with_labels=True, node_size=1000, node_color='red', font_size=12)
plot.show()

cliques = list(nx.find_cliques(model))
print("Clicile: ", cliques)
```
